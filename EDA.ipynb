```python
import pandas as pd
df = pd.read_csv("titanic_cleaned.csv")
df.head()
```

```python

print(df.describe(include='all'))
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
df['Age'].hist(bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

df['Fare'].hist(bins=20, edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()

# Boxplots
sns.boxplot(y=df['Age'])
plt.title('Boxplot - Age')
plt.show()

sns.boxplot(y=df['Fare'])
plt.title('Boxplot - Fare')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Pairplot
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()
```

```python
# Survival by Pclass
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()

# Survival by Gender
sns.barplot(x='Sex_male', y='Survived', data=df)
plt.title('Survival Rate by Gender (1 = Male)')
plt.ylabel('Survival Probability')
plt.show()
```

```python
import plotly.express as px

fig = px.histogram(df, x='Age', color='Survived', barmode='overlay',
                   title='Age vs Survival', nbins=30)
fig.show()

fig = px.box(df, x='Pclass', y='Fare', color='Survived',
             title='Fare by Class and Survival')
fig.show()
```
